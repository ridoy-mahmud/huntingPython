```python
x =[1,2,3]
y =[3,6,7]
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y)
plt.title("Here is the basic title", fontdict={'fontname': 'Comic Sans MS','fontsize':20})
plt.grid()
plt.show()
```

```python
from matplotlib.pyplot import xlim, xticks, yticks


plt.style.use('_mpl-gallery')

# make data:
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

#Sub ploting 
fig,ax=plt.subplots()
ax.bar(x,y,width=1,edgecolor="white", linewidth=0.7)
ax.set(xlim=(0,8),xticks=np.arange(0,8),ylim=(0,8),yticks=np.arange(0,8))
plt.show()
```
